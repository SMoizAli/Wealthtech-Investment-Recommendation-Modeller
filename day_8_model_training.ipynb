```python
import pandas as pd
ap_data=pd.read_csv("../data/day_7.csv")
ap_data.head()
```

```python
ap_data.columns
```

```python
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=ap_data.index,
    y=ap_data['Cumulative_Market_Return'],
    mode='lines',
    name='Market Return'
))

fig.add_trace(go.Scatter(
    x=ap_data.index,
    y=ap_data['Cumulative_Strategy_Return'],
    mode='lines',
    name='Strategy Return'
))

fig.update_layout(
    title='Strategy vs Market Performance',
    xaxis_title = 'Cumulative Return',
    yaxis_title='Cumulative Return',
    template='plotly_white'
)

fig.show()
```

```python
final_strategy = ap_data['Cumulative_Strategy_Return'].iloc[-1]

final_market = ap_data['Cumulative_Market_Return'].iloc[-1]

print("Final Strategy Return : " , final_strategy)
print("Final Market Return : " , final_market)
```
