```python
import pandas as pd 
ap_data=pd.read_csv("../data/appl_day5_behavioral_analysis.csv",index_col='Date',parse_dates=True)
ap_data.head()
```

```python
ap_data.columns
```

```python
ap_data['Recommendataion'].value_counts()
```

```python
def recommand(x):
    if x['Investor_Bias'] == 'Fear' and x['Performance'] == 'Good' :
        return 'Buy'
    elif x['Investor_Bias'] == 'Greed' and x['Performance'] == 'Poor' :
        return 'Sell'
    else : 
        return 'Hold'
    
ap_data['Recommendation']=ap_data.apply(recommand,axis=1)
ap_data[['Investor_Bias','Performance','Recommendation']].head(10)
```

```python
ap_data['Recommendation'].value_counts()
```

```python
import plotly.graph_objects as go
fig=go.Figure()
fig.add_trace(go.Scatter(
    x=ap_data.index,
    y=ap_data['Close'],
    mode='lines',
    name='Price'
))

buy_data=ap_data[ap_data['Recommendation']== 'Hold']

fig.add_trace(go.Scatter(
    x=buy_data.index,
    y=buy_data['Close'],
    mode='markers',
    name='Buy',
    marker=dict(color='green',size=6)
))

hold_data=ap_data[ap_data['Recommendation']=='Hold']

fig.add_trace(go.Scatter(
    x=hold_data.index,
    y=hold_data['Close'],
    mode='markers',
    name='Hold',
    marker=dict(color='Blue',size=3)
))

fig.update_layout(
    title =  'Trading Recommendation System',
    xaxis_title = 'Date',
    yaxis_title='Price'
)

fig.show()
```

```python
ap_data.to_csv('../data/appl_day6_trading_decision.csv')
```
